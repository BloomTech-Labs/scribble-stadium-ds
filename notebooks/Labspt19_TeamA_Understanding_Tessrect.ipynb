```python
from PIL import Image
import pytesseract
import numpy as np
import cv2
```

```python
filename = 'img2.jpg'
img1 = np.array(Image.open(filename))
text = pytesseract.image_to_string(img1)
```

```python
print(text)
```

```python
filename = 'img3.jpg'
img3 = np.array(Image.open(filename))
text = pytesseract.image_to_string(img3)
```

```python
print(text)
```

```python
norm_img = np.zeros((img3.shape[0], img3.shape[1]))
img = cv2.normalize(img3, norm_img, 0, 255, cv2.NORM_MINMAX)
img = cv2.threshold(img3, 100, 255, cv2.THRESH_BINARY)[1]
img = cv2.GaussianBlur(img3, (1, 1), 0)
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#img = mpimg.imread('/content/img3.jpg')
imgplot = plt.imshow(img)
plt.show();
```

```python
text = pytesseract.image_to_string(img)
print(text)
```
